```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```
data =[2, 8, 0, 5, 10, 15, 50, 48, 1]
```

**1. Boxplot**

```
sns.boxplot(x=data)
plt.title('Box Plot')
plt.xlabel('Value')
plt.show()
```

**2. Barchart**

```
less_than_10 = sum(1 for x in data if x < 10)
greater_than_10 = sum(1 for x in data if x > 10)

categories = ['<10', '>10']
counts = [less_than_10, greater_than_10]

sns.barplot(x=categories, y=counts, palette='viridis')
plt.title('Count of Value')
plt.xlabel('Category')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**3. Z-Score Normalization**

```
scaler = stats.zscore(data)
print(scaler)
```
