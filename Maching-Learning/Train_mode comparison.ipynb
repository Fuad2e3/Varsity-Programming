Lab Report 1: Preprocessing

Objective: Perform preprocessing on a Bangla SMS Spam Dataset.

Dataset: https://raw.githubusercontent.com/Ataullha/Machine-Learning-Lab/refs/heads/main/bangla_spam.csv

Steps: Data cleaning, punctuation removal, stopword removal, tokenization, stemming.
Result: Clean dataset ready for training.
Conclusion: Preprocessing improves model performance.

```python
import numpy as np
import pandas as pd
```

```python
Spam = pd.read_csv('/content/Spam/bangla_spam.csv')
```

```python
Spam.head()
```

# **Data Cleaning**

```python
Spam.isnull().sum()
```

```python
Spam.duplicated().sum()
```

```python
Spam = Spam.drop_duplicates()
Spam
```

# **Punctuation Removal**

```python
Spam['text'] = Spam['text'].str.replace('[.,!?:;]', '')
Spam
```

# **Stopword Removal**

```python
stopwords = ['আমি','তুমি','এই','ওই','সে','এটা']
pattern = '|'.join(stopwords)
Spam['text'] = Spam['text'].str.replace(pattern, '')
Spam
```

# **Tokenization**

```python
Spam['tokens'] = Spam['text'].str.split()

Spam
```

# **Stemming**

```python
Spam['text'] = Spam['text'].str[:-1]
Spam
```

Lab Report 2: Machine Learning

Objective: Apply Machine Learning model for classification.

Dataset: Same dataset as Lab Report 01.

Model: Naive Bayes.

Features: TF-IDF.

Result: Good accuracy achieved.

Conclusion: ML models work well with proper features.

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.metrics import accuracy_score
```

```python
x = Spam['text']
y = Spam['type']
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size = 0.2
)
```

# **TF--IDF (Multaple word to number convert)**

```python
Number_convert = TfidfVectorizer(stop_words = stopwords)

# Run the algorithm
Number_convert.fit(x_train)

# Transfrom
xtr_scald = Number_convert.transform(x_train)
xts_scald = Number_convert.transform(x_test)
```

# **TF--IDF (word to number convert)**

```python
le = LabelEncoder()

le.fit(y_train)

# Transfrom
ytr_scald = le.transform(y_train)
yts_scald = le.transform(y_test)
```

# **Model Apply**

```python
nb_model = MultinomialNB()

# run
nb_model.fit(xtr_scald, ytr_scald)
```

# **Score**

```python
y_pred1 = nb_model.predict(xts_scald)

accuracy_score(yts_scald, y_pred1)*100
```

# ***ANN***

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

# **Model Apply**

```python
ann_model = MLPClassifier()

# run
ann_model.fit(xtr_scald, ytr_scald)
```

# **Score**

```python
y_pred1 = ann_model.predict(xts_scald)

accuracy_score(yts_scald, y_pred1)*100
```
